```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

from sailboat_gym import SailboatLSAEnv
```

```python
best_n_envs = pickle.load(open('../output/pkl/best_n_envs.pkl', 'rb'))
time_per_step = np.array(list(best_n_envs.values()))
```

```python
plt.plot(list(best_n_envs.keys()), time_per_step * 1000)
plt.xlabel('n_envs')
plt.ylabel('time per step (ms)');
```

```python
plt.plot(list(best_n_envs.keys()), 1/time_per_step)
plt.xlabel('n_envs')
plt.ylabel('steps per second');
```

```python
plt.plot(list(best_n_envs.keys()), (1/SailboatLSAEnv.SIM_RATE)/time_per_step)
plt.xlabel('n_envs')
plt.ylabel('simulation factor');
```

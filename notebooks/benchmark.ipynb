```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

from sailboat_gym import SailboatLSAEnv
```

```python
def benchmark(time_per_n_envs):
    time_per_step = np.array(list(time_per_n_envs.values())) / 1000 # s
    n_envs = list(time_per_n_envs.keys())
    plt.subplots(figsize=(3,3), dpi=200)
    plt.plot(n_envs, time_per_step * 1000)
    plt.xlabel('n_envs')
    plt.ylabel('time per step (ms)')
    plt.show()

    plt.subplots(figsize=(3,3), dpi=200)
    plt.plot(n_envs, 1/time_per_step)
    plt.xlabel('n_envs')
    plt.ylabel('steps per second')
    plt.show()

    plt.subplots(figsize=(3,3), dpi=200)
    plt.plot(n_envs, (1/SailboatLSAEnv.SIM_RATE)/time_per_step)
    plt.xlabel('n_envs')
    plt.ylabel('simulation factor')
    plt.show()
```
